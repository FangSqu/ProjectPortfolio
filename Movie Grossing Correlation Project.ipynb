# Company Corrlations to Gross

```python
pd.set_option('display.max.row', None)
```

```python
print (df['company'].unique())
```

```python
df_comp = df.groupby('company')
```

```python
df_comp.mean(numeric_only = True).sort_values(by = 'gross', ascending = False). head(10)
```

```python
df = df.dropna()
```

```python
df_numerized = df

for col_name in df_numerized:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized


# Here, the For Loop goes through each column, check for the dtype. if dtype of the column is an object, 
#then it's changed to a 'category'. Then the values were assigned with a 'category code'
```

```python
df_numerized.shape
```

```python
plt.figure(figsize = (20, 10))
corr_mtrx = df_numerized.corr(numeric_only = True)

plt.title('Correlation Matrix Heatmap on Numeric Features of Movie Performance')

sns.heatmap(corr_mtrx, annot =True)
```

```python
# Conclude that the 'company' name has very little correlation with 'gross' (0.15)
# Seems like the factors that has some significant correlation with 'gross' are 'votes'(0.61) and 'budget'(0.74)
```
